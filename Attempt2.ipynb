```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Import the API key(s)
from config import geoapify_key
```

```python
#read in state center lat and lons
state_centers = pd.read_csv('resources/states_centers.csv')
#read in zip code CSV to retrieve state abbreviations for later access
zip_df = pd.read_csv('resources/us_zips.csv', delimiter=";")
#Drop columns other than Zip Code and State, then sort by State while dropping the original index and replacing it with Official State Name
states_clean = zip_df.loc[:,['Official USPS State Code','Official State Name']].sort_values('Official State Name').drop_duplicates().reset_index(drop=True)
#rename for merge
states_abbrev = states_clean.rename(columns={'Official State Name':'state',
                                            'Official USPS State Code':'state_abbrev'})
#Preview dataframe of state abbrev to full name converting dataframe
merged_states = state_centers.merge(states_abbrev,'inner')
merged_states
#now have a clean df of each states center and abbreviation
```

```python
api_sport_list = 'sport,activity.sport_club,commercial.outdoor_and_sport,commercial.outdoor_and_sport.ski,commercial.outdoor_and_sport.diving,commercial.outdoor_and_sport.bicycle,commercial.outdoor_and_sport.golf,building.sport'
base_url = 'https://api.geoapify.com/v2/places'
params={
    #'filter':'circle:-98.35,39.50,4506163'
    #'filter':'circle:-98.350231,39.508968,500000',
    'apiKey': geoapify_key,
    'categories': api_sport_list
}
gym_list = []
for index, row in merged_states.iterrows():
    params['filter'] = 'circle:'+ str(row['longitude']) + ',' + str(row['latitude']) + ',' + str(644542.27)
    response = requests.get(base_url,params).json()
    try:
        list_of_features = response['features']
        gym_list.append(len(list_of_features))
    except:
        gym_list.append(np.nan)
print(gym_list)  
```

```python
api_food_list = 'catering.fast_food'
base_url = 'https://api.geoapify.com/v2/places'
params={
    #'filter':'circle:-98.35,39.50,4506163'
    #'filter':'circle:-98.350231,39.508968,500000',
    'apiKey': geoapify_key,
    'categories': api_food_list,
    'limit':500
}
food_list = []
for index, row in merged_states.iterrows():
    params['filter'] = 'circle:'+ str(row['longitude']) + ',' + str(row['latitude']) + ',' + str(644542.27)
    response = requests.get(base_url,params).json()
    print(requests.get(base_url,params).url)
    try:
        list_of_features = response['features']
        food_list.append(len(list_of_features))
    except:
        food_list.append(np.nan)
print(food_list)  
```
